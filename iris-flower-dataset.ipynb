```python
df = pd.read_csv("../input/iris/Iris.csv")
```

```python
df = df.drop('Id', axis=1)
```

```python
X = np.array(df.drop('Species', axis=1))
```

```python
y = df['Species']

y = y.replace(to_replace='Iris-setosa', value='1')
y = y.replace(to_replace='Iris-versicolor', value='2')
y = y.replace(to_replace='Iris-virginica', value='3')
```

```python
import sklearn
from sklearn.neighbors import KNeighborsClassifier as KNN
from sklearn.model_selection import train_test_split as tts
```

```python
y = np.array(y)
```

```python
X = X/100
```

```python
train_X, test_X, train_y, test_y = tts(X, y, test_size=0.2, random_state=42)
```

```python
model = KNN(n_neighbors=5, algorithm='auto')
```

```python
model.fit(train_X, train_y)
```

```python
from sklearn import metrics
from sklearn.metrics import accuracy_score
```

```python
model.score(train_X, train_y)
```

```python
model.score(test_X, test_y)
```
